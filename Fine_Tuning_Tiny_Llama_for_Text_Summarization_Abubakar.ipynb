```python
def convert_examples_to_features(example_batch):
    # Tokenize the dialogue
    input_encodings = tokenizer(
        example_batch['dialogue'],
        padding="max_length",  # Pad to max_length
        max_length=1024,  # Same max_length for input
        truncation=True,
    )

    # Tokenize the summary (as target)
    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(
            example_batch['summary'],
            padding="max_length",  # Pad to max_length
            max_length=1024,  # Same max_length for target
            truncation=True,
        )

    return {
        'input_ids': input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched=True)
```

```python
# Data Collator with Padding:

from transformers import DataCollatorForSeq2Seq, AutoModelForSeq2SeqLM

# Create the collator
seq2seq_data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,  # Your tokenizer
    model=model_lama,       # Your model
    padding=True,         # Pad sequences to the same length
    max_length=1024,     # Specify the maximum sequence length for padding
)
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='./finetuned-dialogue-summary',
    num_train_epochs=1,
    warmup_steps=500,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    weight_decay=0.01,
    logging_steps=10,
    eval_strategy='steps',
    eval_steps=500,
    save_steps=1e6,
    gradient_accumulation_steps=8,  # Reduced gradient accumulation steps
    fp16=True,  # Enable mixed precision training
    gradient_checkpointing=True, # Enable gradient checkpointing
    report_to="none"
)
```

```python
trainer = Trainer(model=model_lama, args=trainer_args,
                  data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt["test"],
                  eval_dataset=dataset_samsum_pt["validation"])
```

```python
# Fine-tune the model
trainer.train()
```

```python
### Customize before proceeding.

# Save the fine-tuned model
trainer.save_model("./finetuned-dialogue-summary")

# Load the fine-tuned model and tokenizer
model_path = "./finetuned-dialogue-summary"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSeq2SeqLM.from_pretrained(model_path)

# Create a text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, device=device)

# Define a function to generate summaries
def generate_summary(dialogue):
    messages = [{"role": "user", "content": dialogue}]
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    summary = generator(prompt, max_new_tokens=128, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)[0]["generated_text"]
    return summary

# Example usage
dialogue = dataset_samsum["test"][0]["dialogue"]
summary = generate_summary(dialogue)
print(f"Dialogue:\n{dialogue}\n\nSummary:\n{summary}")
```
